```python
## Take some time and think about data

## Median over mean --> Bcz mean can get affected by outliers and median is robust to the outliers
## Here we wont apply median to overall Review Rating, instead we will apply median for Review Rating for each category
## Bcz cloth can have different rating and shoes can have complete different

## We will find median within each category and then place this value in missing value place in each category
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
## Now rename the column names as it contains spaces and upper case -->> Snake casing
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

## Feature engineering
- Create a new column : age_group
- create purchase_frequency_days

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)  ## qcut() --->> Cuts the age into 4 parts 
```

```python
df[['age','age_group']].head(10)
```

```python
## NOw its very easy to analyse the patterns accross different age groups
```

```python
# Create a column 'purchase_frequency_days'
## We have a column 'frequency_of_purchases' but its in a text form, it makes analysis hard, for metrics or compare cust
df['frequency_of_purchases']
```

```python
## So we will change this to numbers
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(5)
```

```python
## Both columns look like same -->> bcz if promocode is used the discount is applied but few companies gives discount without promocode
## So the question is do we need these columns or do both columns have same value, or 1 of the column is redundant
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
## The value came to be True --->> Means both columns have same values
## Drop promo_code_used column

df = df.drop(columns = 'promo_code_used', axis = 1)
```
